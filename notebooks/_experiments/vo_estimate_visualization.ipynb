```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
vo_estimates_euroc = np.load('/Volumes/Data_EXT/data/workspaces/sensor_fusion/notebooks/_experiments/vo_estimates_2d3d_01_v2.npz', allow_pickle=True)
```

```python
estimated_position = vo_estimates_euroc['estimated']
estimated_pose = vo_estimates_euroc['estimated_pose']
ground_truth = vo_estimates_euroc['ground_truth']
```

```python
vo_position = estimated_pose.reshape(-1, 3, 4)[:, :3, 3]
```

```python
%matplotlib widget
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.plot(vo_position[:, 0], vo_position[:, 1], vo_position[:, 2], 'b-', label='Estimated Position')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Trajectory: Estimated vs Ground Truth')
ax.legend()

plt.show()
```

```python
plt.close()
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.plot(ground_truth[:, 0], ground_truth[:, 1], ground_truth[:, 2], 'b-', label='Estimated Position')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Trajectory: Estimated vs Ground Truth')
ax.legend()

plt.show()
```

```python
plt.close()
```
