## Assumption

Since Unscented Kalman Filter has a recursive process meaning that if error between estimates and measured vector are corrected properly in measurement update step, the mean and its error covariance matrix continuously grow over time, causing the "4-th leading minor of the array is not positive definite" error since the covariance matrix P breaks a convention that the matrix has to be positive-definite meaning that all the eigenvalues need to be positive or zero.
