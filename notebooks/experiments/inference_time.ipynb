This notebook is created to compare an inference time among multiple computer, which includes:

- M1 Macbook Pro
- Raspberry Pi4.

```python
import sys
import numpy as np
import pandas as pd
src_dir = "../../src"
if src_dir not in sys.path:
    sys.path.append(src_dir)
```

```python
results_header = pd.MultiIndex.from_product([['Setup1 (IMU, VO)','Setup2(IMU, VO+GPS)', 'Setup3(INS, VO+INS)'],
                                             ['EKF', 'UKF', 'PF', 'EnKF', 'CKF']],
                                            names=['Setups','Filter types'])
indices = ["M1 Macbook Pro", "Raspberry pi 4"]
```

```python
m1_macbook_pro = np.load("../../exports/inference_time/m1_macbook_pro/inference_time.npy", allow_pickle=True)
raspberry_pi4_ubuntu = np.load("../../exports/inference_time/raspberry_pi4_ubuntu/inference_time.npy", allow_pickle=True)

data = np.concatenate([
    m1_macbook_pro.reshape(1, -1),
    raspberry_pi4_ubuntu.reshape(1, -1),
], axis=0)
data
```

```python
df = pd.DataFrame(
    data, 
    index=indices, 
    columns=results_header
)
```

## Inference time in milliseconds

```python
df
```

In the table, Extended Kalman Filter has the samllest inference time over the other filter methods in all the setups.
In the case of using a powerfull computer such as Macbook Pro, all filter methos have acceptable inference time to employ. However, a SBC, such as Raspberry pi 4, it is feasible to avoid the Particle Filter (PF) since it inference time exceeds 500 milliseconds when both GPS and VO are taken into account in the measurement update step, performed in the Setup2 and Setup3. Other than the PF, any filter methods are acceptable to embed.
